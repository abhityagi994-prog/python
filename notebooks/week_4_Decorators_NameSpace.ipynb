```python
#Namespaces
#A namespace is a space that holds names(identifiers).Programmatically speaking, 
#namespaces are dictionary of identifiers(keys) and their objects(values)

#There are 4 types of namespaces:

#Builtin Namespace
#Global Namespace
#Enclosing Namespace
#Local Namespace
```

```python
#Scope and LEGB Rule
#A scope is a textual region of a Python program where a namespace is directly accessible.

#The interpreter searches for a name from the inside out, looking in the local, enclosing, global, 
#and finally the built-in scope. If the interpreter doesn’t find the name in any of these locations, 
#then Python raises a NameError exception.
```

```python
# local and global
# global var
a = 2

def temp():
  # local var
  b = 3
  print(b)

temp()
print(a)
```

```python
# local and global -> same name
a = 2

def temp():
  # local var
  a = 3
  print(b)

temp()
print(a)
```

```python
# local and global -> local does not have but global has
a = 2

def temp():
  # local var
  print(a)

temp()
print(a)
```

```python
# local and global -> editing global
a = 2

def temp():
  # local var
  a += 1
  print(a)

temp()
print(a)

#The above code will throw an error because. The function doesn't has its own local varibale and "a" local
#variable is trying to change the value of global varibale. Which is not allowed. 
```

```python
#We can create the global varible using the local namespace

def temp():
    global a
    a=1
    print(a)

temp()
print(a)

#As we can see from the above example that if we create the globale varibale inside the function. It will
#Create a new global variable as well.
```

```python
#Very Important Example - 

L=[1,2,3]
max(L)
def max():
    print("Hello")

max(L)

#The above code is throwing an error because we have override the built-in function max with our own function. 
#According to the LEGB rule -> First python will check do we have any max in Local, after it will check in 
#Enclosure, and finally to Global and Built-in. But as we have max in global, so our function max will get
#executed instead of the buit-in function
```

```python
#Enclosure Scope - It is usually found in nested loops. 
# Enclosing scope
def outer(): 
    a=1
    def inner():
        print(a)
    inner()
print('outer function')


outer()
print('main program')
```

```python
# nonlocal keyword
def outer():
  a = 1
  def inner():
    a += 1
    print('inner',a)
  inner()
  print('outer',a)


outer()
print('main program')

#The above program will throw an error. Because the inner scope does not have its own local variable. 
#But it is trying to change the value of enclosure scope variable a=1, which is not allowed.
#The concept is same as when we don't have the value in local varibale but trying to change the
#value of global variable.
```

```python
# nonlocal keyword
def outer():
  a = 1
  def inner():
    nonlocal a
    a += 1
    print('inner',a)
  inner()
  print('outer',a)


outer()
print('main program')

#Using the nonlocal variable we can also change the value of Enclosure scope variable.
```

```python
#Decorators
#A decorator in python is a function that receives another function as input and adds 
#some functionality(decoration) to and it and returns it.

#This can happen only because python functions are 1st class citizens.  -->Important
#First class citizens means we can perform operations of functions such as we can store them in another
#datatype or aliasing or we can delete them.

#There are 2 types of decorators available in python

#Built in decorators like @staticmethod, @classmethod, @abstractmethod and @property etc
#User defined decorators that we programmers can create according to our needs
```

```python
#Example of how decorators work - 

def my_decorator(func):
    def wrapper():
        print("********************")
        func()
        print("********************")
    return wrapper

def name():
    print("Hello Name!")

a=my_decorator(name)
a()
```

```python
#Another simple example fo decorator

def outer():
    a=5
    def inner():
        print("***************")
        print(a)
        print("***************")
    return inner
    

b=outer()
b()
```

```python
#Another short way to write a decorator is:

def my_decorator(func):
    def wrapper():
        print("********************")
        func()
        print("********************")
    return wrapper

@my_decorator
def name():
    print("Hello Name!")

name()
```

```python
#In built decorator @classmethod

class Person:
    def __init__(self,fname,lname):
        self.lname=lname
        self.fname=fname

    def __str__(self):
        return ('{}'+" "+'{}').format(self.fname,self.lname)
        
    @classmethod
    def reference(class_,fname,lname):
        return class_(fname,lname)


pr=Person("Lord","King")
print(pr)
```

```python
pr1=Person.reference("Lord","King")
print(pr1)
```

```python
#Custom decorator

import time
def time_decorator(func):
    def wrapper():
        start=time.time()
        func()
        stop=time.time()
        print(f"Total time taken by the function {func.__name__} is: ",stop-start)  #func.__name__ using this
    return wrapper                                                                  #we can get tje function name


@time_decorator
def hello():
    time.sleep(2)
    print("I simply a function!!")

@time_decorator
def display():
    time.sleep(4)
    print("I am another simple function")

hello()
display()
```

```python
#Custom decorator

import time
def time_decorator(func):
    def wrapper(*args):
        start=time.time()
        func(*args)
        stop=time.time()
        print(f"Total time taken by the function {func.__name__} is: ",stop-start)
    return wrapper



#@time_decorator
def square(a):
    time.sleep(1)
    print(a**2)
    print("I am another simple function")

a=time_decorator(square)
a(2)
```

```python
@time_decorator
def cube(a):
    time.sleep(1)
    print(a**3)
    print("I am another simple function")

cube(3)
```

```python
#A really Good Example


def powerful_decorator(data_type):
    def outer_wrapper(func):
        def inner_wrapper(*a):
            if type(*a)==data_type:
                func(*a)
            else:
                raise Exception("Nahi Chalega")
        return inner_wrapper
    return outer_wrapper

@powerful_decorator(int)
def cube(num):
    print(num**3)

try:
    cube(5)
except Exception as e:
    print(e)
```

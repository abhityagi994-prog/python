## Generators 
- Generators are the simple way of creating an iterators.

```python
#Creating a simple Generator function

def gen():
    yield "Hello"
    yield "How are you"
    yield "I am fine thank you"

g=gen()   #Creating the object of generator class. Simple "gen" is a function. "gen()" is a generator class object.
print(id(gen()))   #2nd Object of generator class
print(type(gen()))
print(id(g))       #1st object of generator class
print(type(g))

#Hence, the ids are also different for different objects
```

```python
print(next(g))
print(next(g))
print(next(g))
#print(next(g))

#Important thing to notice is when we are calling the generator function, it gives us the generator object. And if we pass that object through 
#next, it will print the yield statement. Until StopIteration error is raised
```

```python
for i in g:
    print(i)
```

#### Difference b/w Generator function and normal return function
- Refer the notes

```python
#Printing the squares 

def square(num):
    for i in range(1,num+1):
        yield i**2

gen=square(10)
print(next(gen))
print(next(gen))
print(next(gen))

for i in gen:
    print(i)  #As we can see in the middle we started the for loop, but still generator function remembers what as the last item, so it continue
              #after that.
```

#### Creating our own range function using the Generator

```python
def our_range(start,stop):
    for i in range(start,stop):
        yield i

for i in our_range(1,10):
    print(i)
```

```python
r=our_range(1,10)
print(next(r))
print(next(r))
print(next(r))
```

#### Generator Expression

```python
gen=(i for i in range(1,10))

print(next(gen))
print(next(gen))
for i in gen:
    print(i)
```

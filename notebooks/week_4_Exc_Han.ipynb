```python
#More advance way to show the error
try:
    with open("demo.txt","r") as f:
        print(f.m())
except Exception as e:    
    print(e)
    #OR
    print(e.with_traceback)

#Now, using the traceback we can know the name of error we are having. So, we can replace the 
#exception with the name of the error.
```

```python
try:
    with open("demo.txt","r") as f:
        print(f.m())
except FileNotFoundError:   #As we can see for the particular type of error we have our except block.
    print("File not found") 
except AttributeError:   #For attribute error also we have an exception block.
    print("File does not have such an attribute")  
except Exception as e:
    print(e)
```

```python
#Else statement

try:
    f=open("demo.txt","r")
except FileNotFoundError:   #As we can see for the particular type of error we have our except block.
    print("File not found") 
except AttributeError:   #For attribute error also we have an exception block.
    print("File does not have such an attribute")  
except Exception as e:
    print(e)
else:
    print(f.read())

#In try and except, else statement will get executed at the end after checking for except blocks. 
#Either except will get executed or else will get executed
```

```python
try:
    f=open("demo.txt","r")
except FileNotFoundError:
    print("File is not in the directory")
except AttributeError:
    print("File has not such attribute")
except Exception as e:
    print(e)
else:
    print(f.read())
finally:
    print("Irrespective of anything I will execute") 

#Finally doesn't care whether except block will get execute or else. But finally will going to be execute
```

```python
class Bank:
    def __init__(self,balance):
        self.balance=balance

    def withdraw(self,amount):
        if amount<0:
            raise Exception("Amount is in negative")    #Here exception is the exception class
        elif amount>self.balance:
            raise Exception("Amount is bigger than balance \U0001F608")
        else:
            self.balance=self.balance-amount

bnk=Bank(15000)
```

```python
bnk.balance
```

```python
try:
    bnk.withdraw(1000)
except Exception as e:
    print(e)
else:
    print(bnk.balance)
```

```python
class MyError(Exception):
    def __init__(self,message):
        print(message)
        #pass
```

```python
class Bank:
    def __init__(self,balance):
        self.balance=balance

    def withdraw(self,amount):
        if amount<0:
            raise MyError("Amount is in negative \U0001F611")
        elif amount >self.balance:
            raise MyError("Amount is greater than the balance \U0001F608")
        else:
            self.balance=self.balance-amount

bnkk=Bank(20000)
```

```python
try:
    bnkk.withdraw(-100000)
except MyError:  #Or Exception
    pass      #becuase we are already printing the error using our custom class
else:
    print(bnkk.balance)
```

```python
#Main question is why do we need custom class errors?
#1. Full control
#2. Application based Errors
#It means if we have our own logic for our particular application, we can create our custom errors.
```

```python
class SecurityError(Exception):
    def __init__(self,message):
        print(message)

    def logout(self):    #With the help of custom error class, we were able to create a new logic of logout
                         #This is the benefit of custom class erros, that we can create our own logic   
        print("Logged out from all the devices")

class Google:
    def __init__(self,name,email,pass_,device):
        self.name=name
        self.email=email
        self.pass_=pass_
        self.device=device

    def login(self,email,pass_,device):
        if device!=self.device:
            raise SecurityError("High Alert!!!!! ")
        elif email==self.email and pass_==self.pass_ and device==self.device:
            print("Logged IN")
        
```

```python
obj=Google("Holy","holy@12",123,"android")
```

```python
try:
    obj.login("holy@12",123,"android")
except SecurityError as e:
    e.logout()
else:
    print(f"{obj.email}, {obj.device}, {obj.name}")
```

#### Iteration is a general term we use for taking out the items or going over the items one by one. Whenever we are using the
#### loop implicitly or explicitly we are performing an iteration.

```python
L=[i for i in range(1,101)]  #We have created the List of 100 items using the for loop. The process through which these
print(L)                     #these items are generated is called Interation. 
```

#### Iterator is an object that allow the programmer to transverse through the sequence of data, without storing all the
#### data into the memory

```python
import sys

print(sys.getsizeof(L))  #This is the memory our list is taking.

#Now we will again going to print the numbers from 0 to 100, but this time using the range function only.

x=range(1,101)
for i in x:
    print(i)
```

```python
print(sys.getsizeof(L))
print(sys.getsizeof(x))

#As we can see that there is a very big difference of memory consumption. Because range function is an iterator which is going through each item
#one by one with without storing all the items in the memory. 
```

#### An Iterable is any object on which an iteration can be performed. It also generates an iterator when passes through 
#### the iter() method.

```python
L=[1,2,3]

#Method - 1 to check whether an object is iterable is run a loop on it. If we can, then the object is iterable or not is to run a loop.

for i in L:
    print(i)
```

```python
a=2
for i in a:
    print(i)

#As we can see, we cannot run a loop on it. Hence, 2 is not an iterable
```

```python
#Method -2 to find out which object is iterable

print(dir(L))
#As we can see, in dir we have __iter__. Hence, it ,means L is an Iterable

print("\n")

print(dir(a))
#As we can see there is no __iter__ in dir for a. Hence, the obejct is not iterable
```

## Important Points

1. An Iterator is also an Iterable.
2. All Iterables cannot be a Iterators.

```python
L=[1,2,3]
print(dir(L))   #L is an iterable, but not an iterator because in dir(L) has __iter__, but not __next__. 
```

#### How to generate the iterator using the iterable?

```python
L=[1,2,3]
print(iter(L))   #Passing the Iterable through the iter() method can generate the Iterator

#Now, Importanrt thing to remember is although the List is iterable but it is not an iterator. Because L store all the items in the memory
#Unlike the range function which takes the item one by one and keep only the num on which its iterator currently is.
```

```python
t=(1,2,3)

print(dir(t))
```

```python
z=iter(t)
print(dir(z))
#As we can see, z has both the __iter__ and __next. Hence, it is an iterator.
```

```python
#For range function

x=range(1,5)
print(dir(x))
print("\n")
y=iter(x)
print(dir(y))
```

### Creating a for loop 

```python
def _my_for_loop(iterable):
    iterator=iter(iterable)
    while True:
        try:
            print(next(iterator))
        except StopIteration:
            break


a=[1,2,3]
b=range(1,10)
c=(4,1,2)      
_my_for_loop(a)
print("\n")
_my_for_loop(b)
print("\n")
_my_for_loop(c)
```

#### **An Important Point**

```python
L=[33,44,55]  #An Iterable, but not an iterator
x=iter(L)     #An Iterator of an Iterable using iter()
print(x)
print(id(x))  #This is the id of our iterator 
print("\n")
y=iter(x)     #An iterator x
print(y)      #Here y is the iterator of an iterator
print(id(y))  #Memory address of iterator y

#Main concept - Why we are doing it, is to explain that if we do iter() on a iterator, it will give us another iterator which is that iterator itself.
```

### My own range function

```python
class own_range:

    def __init__(self,start,stop):
        self.start=0
        self.stop=stop

    def __iter__(self):
        return own_iterator(self)   #Its an object of Iterator class. As we alredy know that if we pass iterable into iter() method. We get an 
                                    #iterator object.
```

```python
class own_iterator:

    def __init__(self,iterable:own_range):
        self.iterator=iterable

    def __iter__(self):
        return self

    def __next__(self):
        if self.iterator.start>=self.iterator.stop:
            raise StopIteration
        else:
            current=self.iterator.start
            self.iterator.start+=1
            return current
        
        
```

```python
for i in own_range(1,11):
    print(i)
```

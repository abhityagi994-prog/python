```python
#Seek help us in positioning our file curser. With the help of seek we can position or
#file pointer at any location.

with open("demo.txt","r") as f:
    print(f.read(18))
    print(f.tell())
    f.seek(0)     #Using the seek to bring the position of file curser to 0th position.
    print(f.tell())
    print(f.read(18))
```

```python
#Serialization and Deserialization
#· Serialization - process of converting python data types to JSON format
#· Deserialization - process of converting JSON to python data types
```

```python
#Serialization

L=[1,2,3,4,5]

import json

with open("demo.json","w") as f:    #We have to store the file in JSON or else. If we use the file somewhere else
                                    #.txt file might misbehaive.
    json.dump(L,f)      #Dump is a method of json, in which we have to put two arguments. One is the datatype
                        #And another is the file

with open("demo.json","r") as f:
    print(type(f.read()))   #It is showing str becuase it doesn't matter what type of data we store, if we 
                            #call read() it will show str. As python read the text files in str format

with open("demo.json","r") as f:
    print(type(json.load(f)))   #json load actually tells us what type of value is stored in the file.As it gives
                                #back the types of data we stored in file. Deserialization. 

#We could say json.dump() is store and json.load() is restore
```

```python
d={1:11,2:22,3:33}
with open("demo.json","w") as f:
    json.dump(d,f,indent=4)     #Here indent is nothing but just the structure. Even if we removed the indent
                                #nothing will change, just the dictionary will be in straight line.
```

```python
#Deserilization

with open("demo.json","r") as f:
    d=json.load(f)   #data restore
    print(d)
    print(type(d))
```

```python
#How to store the custom object

class Person:
    def __init__(self,name,age,position):
        self.name=name
        self.age=age
        self.position=position


obj=Person("Jacob",99,"Pilot")
```

```python
with open("demo.json","w") as f:
    json.dump(obj,f)

#The above code will throw an error, becuase we cannot store the custome objects like this.
```

```python
def storing_obj(obj):   #This function is receiving an object and checking whether that
                        #that object belong to the Person class or not
    if isinstance(obj,Person):
        return f"Name - {obj.name}, Age - {obj.age}, Position - {obj.position}"

with open("demo.json","w") as f:
    json.dump(obj,f,default=storing_obj)   #default is the perimeter to tell the json how to store the object
```

```python
#The problem with the above method is we have to define the how we want to show our 
#object like in the format of list or dictionary or normal print. 
#But what if I want to use my object as it is. Means I don't want to define how I want to show my object
#instead use it as it is.

#Pickling
#Pickling is the process whereby a Python object hierarchy is converted into a byte stream, and
#unpickling is the inverse operation, whereby a byte stream (from a binary file or bytes-like object) is
#converted back into an object hierarchy.

#In simple words Pickling means we convert the object into the binary form and can transfer it to another file.
#And Unpickling means converting the binary form back to object 
```

```python
class Prsn:
    def __init__(self,fname,lname,age):
        self.fname=fname
        self.lname=lname
        self.age=age

    def get_info(self):
        return f"{self.fname} {self.lname} - {self.age}"

pr1=Prsn("John","Dawn","92")
```

```python
#Pickling

import pickle

with open("demo.pkl","wb") as f:    #for pickling we have to store our custom object in .pkl file 
    pickle.dump(pr1,f)              #and similar to json.dump we have pickle.dump
    
```

```python
#Unpickling

with open("demo.pkl","rb") as f:
    d=pickle.load(f)   #Using load we can unpickle our object and can use it.

d.get_info()   
```

```python
#Pickle VS JSON 

#Pickle is used when we want to store our data in Binary format. And JSON is used when
#we want to store our data in human readable format
```
